```python
# Load the JSON data
file_path = './output/lr_history.json'
with open(file_path, 'r') as file:
    lr_history = json.load(file)

# Extract learning rates and losses
learning_rates = lr_history['lrs']
losses = lr_history['losses']

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(learning_rates, losses, marker='o', linestyle='-')
plt.xscale('log')  # Since learning rate changes exponentially, log scale might be more informative
plt.xlabel('Learning Rate')
plt.ylabel('Loss')
plt.title('Learning Rate vs. Loss')
plt.grid(True)
plt.show()
```

```python
# Load the JSON data
file_path = './output/loss_history.json'
with open(file_path, 'r') as file:
    loss_history = json.load(file)

# Extract learning rates and losses
train_total = loss_history['train_total']
train_lx = loss_history['train_lx']
train_lm = loss_history['train_lm']

val_total = loss_history['val_total']
val_lx = loss_history['val_lx']
val_lm = loss_history['val_lm']

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(train_total, marker='o', linestyle='-')
plt.xlabel('Step')
plt.ylabel('Train Loss')
plt.grid(True)
plt.show()

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(train_lx, marker='o', linestyle='-')
plt.xlabel('Step')
plt.ylabel('Train Lx')
plt.grid(True)
plt.show()

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(train_lm, marker='o', linestyle='-')
plt.xlabel('Step')
plt.ylabel('Train Lm')
plt.grid(True)
plt.show()

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(val_total, marker='o', linestyle='-')
plt.xlabel('Step')
plt.ylabel('Val Loss')
plt.grid(True)
plt.show()

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(val_lx, marker='o', linestyle='-')
plt.xlabel('Step')
plt.ylabel('Val Lx')
plt.grid(True)
plt.show()

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(val_lm, marker='o', linestyle='-')
plt.xlabel('Step')
plt.ylabel('Val Lm')
plt.grid(True)
plt.show()
```
